# 1. 선 그래프 (Line Graph) 그리기
## 1.1 선 그래프(꺽은선 그래프)
- 점과 점을 선으로 연결한 그래프
- 시간의 흐름에 따른 변화를 표현할 때 많이 사용한다. (시계열) // 산점도는 순서 흐름을 시각적으로 파악할 수 없다.
- `plot([x], y)` 
    - 1번인수 : x값(생략가능), 2번인수 y값
    - 인수가 하나인 경우 y 축의 값으로 설정되고 X값은 (0 ~ len(y)-1) 범위로 지정된다.
    - x,y 의 인수는 리스트 형태의 객체들을 넣는다.
        - 리스트
        - 튜플
        - numpy 배열 (ndarray)
        - 판다스 Series
    - x와 y의 size는 같아야 한다.
- 하나의 axes(subplot)에 여러 개의 선 그리기
    - 같은 axes에 plot()를 여러번 실행한다.

```python
plt.figure(figsize = (5,5))
plt.plot([3,4,5])
plt.show()
```

## 1.2 선 스타일
- https://matplotlib.org/3.0.3/gallery/lines_bars_and_markers/line_styles_reference.html

```python
import matplotlib.pyplot as plt
plt.plot(range(1, 20, 5), range(10, 30, 5))
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

x = range(1,  11)
y = pd.Series(range(10, 110, 10)) # 원소별로 계산하려고 Series로 만들어줌.
y2 = y + 5
y3 = y + 10
y4 = y + 15

# print(len(x), len(y)) # x과 y의 size는 같아야 한다.

# figure size 설정 (가로, 세로)
plt.figure(figsize = (7,5))

# (x축값, y축값, [label], [linestyle])
plt.plot(x, y, label = 'y') # linestyle defult
plt.plot(x, y2, label = 'y+5', linestyle = ':')
plt.plot(x, y3, label = 'y+10', linestyle = '-.')
plt.plot(x, y4, label = 'y+15', linestyle = '--')

plt.legend() # 범례

plt.show()
```

## 1.3 선 그래프 활용
- [서울시 연도별 황사 경보발령 현황](http://data.seoul.go.kr/dataList/datasetView.do?infId=OA-2227&srvType=S&serviceKind=1&currentPageNo=1)
- 연도별 관측일수와 황사최대농도의 변화를 그래프로 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/서울시 연도별 황사 경보발령 현황.csv')
df.shape
df

# 최대 농도 컬럼명 바꿔주기
df.rename(columns = {'최대농도(㎍/㎥/시)':'최대농도'}, inplace = True)
```

```python
df
```

```python
# 년도흐름에 따른 최대농도의 변화량.
# x축 : 시간  y축 : 깂 

plt.figure(figsize=(10,6)) # 보통 x축 가로가 길게 y축은 짧게

plt.plot(df['년도'], df['최대농도'])

# axes 제목
plt.title('연도별 황사 최대농도 변화량')

# x,y 축 label
plt.xlabel('측정년도')
plt.ylabel('황사최대농도')

# tick은 y축의 min과 max에 따라 min 근처값 ~ max 근처값 

plt.show()
```

```python
# 년도 별 관측일수의 변화량

plt.figure(figsize = (7,5))
plt.plot(df["년도"], df['관측일수'], c = 'g')

plt.title('년도별 관측일수의 변화량', fontsize='20', c = 'b')
plt.xlabel('측정년도', c = 'red')
plt.ylabel('관측일수')

plt.show()
```

## 최대 농도와 관측일 수의 연도별 변화를 시각화
- ### 하나의 축을 공유하고 두개의 축을 가지는 그래프 그리기
- 값의 범위(Scale)이 다른 두 값과 관련된 선그래프를 한 Axes(subplot)에 그리는 경우
- X축을 공유해 2개의 Y축을 가지는 그래프
    - axes.twinx() 를 이용해 axes를 복사 (객체지향 방식)
- Y축을 공유해 2개의 X축을 가지는 그래프
    - axes.twiny() 를 이용해 axes를 복사

```python
# 한 figure 내에의 다른 axes(subplot) 에 각각 그리기.

fig = plt.figure(figsize= (20, 6))

# figure의 title
fig.suptitle('년도별 각종 변화량', fontsize = 25)

plt.subplot(1,2,1)
plt.plot(df['년도'], df['최대농도'])
plt.title('년도별 최대농도 변화량') # axes 별 title
plt.xlabel('년도')
plt.ylabel('최대농도')

plt.subplot(1,2,2)
plt.plot(df['년도'], df['관측일수'])
plt.title('년도별 관측일수의 변화량') # axes 별 title
plt.xlabel('년도')
plt.ylabel('관측일수')

plt.tight_layout() 
plt.show()
```

```python
# 하나의 axes(subplot)에 두개를 같이 그리기 
# 두 컬럼(변수)의 값의 범위(scale)이 같거나 비슷한 경우 그리는 것이 좋다.

# 만약 지금 처럼 범위 차이가 많이 난다면, 한 축은 공유하고 범위차이 큰 축은 두 개로

plt.figure(figsize=(15,5))

plt.plot(df['년도'], df['최대농도'], label = '최대농도변화량')
plt.plot(df['년도'], df['관측일수'], label = '관측일수변화량')

plt.title('년도별 최대농도와 관측일수 변화량')

plt.xlabel('년도')
plt.ylabel('최대농도/관측일수')

plt.legend() # 범례
plt.show()
```

```python

# 객체지향방식으로 해야함.

fig = plt.figure(figsize=(15, 5))

# axes 
ax1 = fig.add_subplot(1,1,1) #plt.gca() : AxesSubplot 반환. 한 그래프그릴땐 이렇게 해도 됨.
ax2 = ax1.twinx() # ax1과 x를 공유하며 위에 ax2를 합쳐 겹치게 만들어준다.

# plot
ax1.plot(df['년도'], df['최대농도'], label='최대농도', color='blue')
ax2.plot(df['년도'], df['관측일수'], label= '관측일수', color = 'g')

# x,y축 명 지정
ax1.set_xlabel('년도')
ax1.set_ylabel('최대농도', c='r')
ax2.set_ylabel('관측일수', c='r')


# 겹쳐 있으니 둘 중 하나 하던가 아님 figure title로 하던가
ax1.set_title('년도별 최대농도/관측일수 변화량') 


# 현재 두 그래프는 다른 axes. 다른 axes이기에 각각해줘야 함.
# 설정 안하면 ax1 범례에 위에 겹쳐쓰게 된다.

# 범례위치를 지정. loc = '상하위치 좌우위치' 박스 9 구간 셋팅 가능.(자유롭게 불가능)
#                         상하 : lower, center, upper
#                         좌우 : left, center, right
# ax1.legend(loc = 'center') # 하나만 하면 상하 좌우 둘다 가운데
# ax2.legend(loc= 'upper left')

# 자유롭게 지정할 때
ax1.legend(bbox_to_anchor=(1.04,1), loc = 'upper left')
ax2.legend(bbox_to_anchor=(1.04,0.94), loc = 'upper left') 

plt.show()
```

### legend 위치 설정

- Axes 안의 정해지 위치(9곳)에 설정
    - 현재 두 그래프는 다른 axes. 다른 axes이기에 각각해줘야 함.
    - 설정 안하면 ax1 범례에 위에 겹쳐쓰게 된다.
    - 범례위치를 지정. loc = '상하위치 좌우위치' 박스 9 구간 셋팅 가능.(자유롭게 불가능)
    - 상하 : lower, center, upper
    - 좌우 : left, center, right
    - 기본(def) : best(알아서 최적)
    - ax1.legend(loc = 'center') # 하나만 하면 상하 좌우 둘다 가운데
    - ax1.legend(loc= 'upper left')
- 원하는 위치를 지정
    - bbox_to_anchor,  loc  두 개 매개변수 이용
    - bbox_to_anchor : legend box의 Axes상의 위치-(상하, 좌우)
        - (상하, 좌우)
        - 전체 Axes(subplot)기준
            - 상단 : (0,1) (1,1)
            - 하단 : (0,0) (1,0)
    - loc : legend box의 기준위치(기준점이 될 위치) 
    - legend box의 loc을 Axes(Subplot)의 bbox_to_anchor에 위치
        - ax1.legend(bbox_to_anchor=(1,1), loc = 'upper left') 을 주로 쓴다.

```python
# - Axes 안의 정해지 위치(9곳)에 설정

plt.plot([1,2,3], [1,2,3], label = 'test')
plt.legend(loc = 'lower right',fontsize = 20)
plt.show()
```

# 2. 산점도 (Scatter Plot) 그리기
## 2.1 산점도(산포도)
- X와 Y축을 가지는 좌표평면상 관측값들을 점을 찍어 표시하는 그래프
- 변수(Feature)간의 상관성(상관관계. ex. x축에 따라 y축이 증가)이나 관측값들 간의 군집 분류(ex. 두 개의 그룹으로 나눠, 뭉쳐져 있다.)를 확인할 수 있다.
    - cf) line plot는 값의 위치보다 시간, 순서 등 변화 흐름에 따른 값 보고 싶은 것.
- `scatter()` 메소드 사용
    - 1번인수 : x값, 2번인수 y값
    - x와 y값들을 모두 매개변수로 전달해야 한다.
    - x,y 의 인수는 스칼라 실수나 리스트 형태의 객체들을 넣는다.
        - 리스트
        - 튜플
        - numpy 배열 (ndarray)
        - 판다스 Series
    - x와 y의 원소의 수는 같아야 한다.

#### 패턴을 아는 것은 중요하다. -> 패턴이 있어야 예측이 가능하다.

```python
import numpy as np

x = np.arange(1,1000,50) # 1 ~ 1000-1  +50 # range와 동일하게 마지막 인덱스 빼고 50씩
y = np.arange(1,1000,50) 

plt.figure(figsize=(7,7))
plt.scatter(x,y) # x와 y 같은 인덱스의 값끼리 묶어서 점을 찍는다.
plt.show()
```

```python
y2 = np.random.randint(100,300,20)  # 100 ~ 300-1 사이에서 임의의 정수 20개 (난수)를 추출.

plt.scatter(x, y2)
plt.show()

# 상관계수 계산
np.corrcoef(x,y2)

# 이러한 그래프는 변수간의 상관성이 없다라고 말할 수 있다.
```

## 2.2 설정
- marker (마커)
    - marker란 점의 모양을 말하며 미리정의된 값으로 변경할 수있다.
    - scatter() 메소드의 marker 매개변수를 이용해 변경한다. 
    - https://matplotlib.org/stable/api/markers_api.html
- s
    - 마커의 크기
    - 리스트형태(리스트, 튜플)로 지정
- alpha    
    - 하나의 마커에 대한 투명도
    - 0 ~ 1 사이 실수를 지정 (default 1)

```python
plt.figure(figsize=(10,10))

plt.scatter(x=np.random.randint(1,5,30), y=np.random.randint(1,5,30), 
            marker = 'v',
            color = 'red',
            s = 100, # maker size
            alpha = 0.1 # 투명도 0~1
           ) 
# 중복되어 찍혀서 점이 30개로 안 보이는데, 이 정도를 투명도로 알 수 있다.

plt.scatter(x=np.random.randint(5,10,30), y=np.random.randint(5,10,30), marker = '*')

# 군집 분류를 알 수 있는 그래프
```

### 경보 발령횟수와 관측일수 간의 관계
- 경보발령횟수가 많았던 년도는 관측일수도 길었나?

```python
df
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df['경보 발령횟수'], df['관측일수'])

plt.title('경보발령횟수와 관측일수간의 관계')
plt.xlabel('경보발령횟수')
plt.ylabel('관측일수')

# x축 눈금(tick) 재조정 -> xticks
plt.xticks([0,1,2]) # (눈금의 위치, 눈금 label) label 생략 시 눈금의 value(값)이 label이 된다.

plt.yticks([1, 10, 15])

plt.show()
```

#### 두 컬럼(변수) 간의 상관관계(한쪽의 값이 증가할 때 다른 쪽 값은 증가 or 감소 or 관계없음)를 계산

- 정상적 판단 : 산점도(scatter plot)
- 정량적 판단 : 상관계수 계산
    - 상관관계 : 양의 상관관계(비례 증감이 같은 것), 음의 상관관계(반비례 증감이 반대인 것)
    - 상관관계 : -1 ~ +1 실수
        - 음수 : 음의 상관관계, 양수 : 양의 상관관계, 0 : 관계없다.
        -  +-1에 가까울 수록 관계성이 커진다. 0에 가까울수록(퍼질수록) 관계성이 약해진다.
        - 1 ~ 0.7 : 아주 강한 상관관계
        - 0.7 ~ 0.3 : 강한 상관관계
        - 0.3 ~ 0.1 : 약한 상관관계
        - 0.1 ~ 0 : 상관성(관계)이 없다.
   - 계산법 `np.corrcoef()`


```python
# 상관계수 계산 - numpy

import numpy as np

np.corrcoef(df['경보 발령횟수'], df['관측일수'])

# 0.6 : 양의 상관관계(양수) -> 강한 관계성을 가진다. 비례.
```

```python
# 상관계수 계산 - pandas

df[['경보 발령횟수', '관측일수']].corr()
```

# 3. 막대그래프 (Bar plot) 그리기
## 3.1 막대그래프(Bar plot)

- 수량을 막대 형식으로 나타낸 그래프 
    - 수량을 표현할 때 사용. (ex. 범주형컬럼의 value_counts() 각 범주별 수량 갯수 ) 

- 수직막대 : **bar(x, height) 메소드 사용** 
    - X: x값 -> 분류값(대상)
    - height:  막대 높이 ->  개수 수량
        - 수직막대를 주로 사용함.
- 수평막대 : **barh(y, width) 메소드**
    - 1번인수: y값 (수량을 셀 대상)
    - 2번인수: 막대 너비 (수량)
        - 대상의 이름이 길 때 label이 겹치질 수 있을 때 사용.

```python
import matplotlib.pyplot as plt
fruits = ['사과', '참외', '딸기']
counts = [100, 200, 50]  # 사과, 참외, 딸기의 개수
```

```python
# 데이터가 많다면 수량 파악이 어려울 것.
plt.scatter(fruits, counts)
```

- 막대 그리기
- 막대 두께 조절
    - 수평 : width = 0~1 실수, 
    - 수직 : hight = 0~1 실수


```python
# 수직막대 그래프
plt.figure(figsize = (15,7))
plt.subplot(1,2,1)
plt.bar(fruits, counts, width=1)  # (대상 x, 수량 y, width(def 0.8) (1이면 붙는다.)

plt.title('수직막대그래프')
plt.xlabel('과일')
plt.ylabel('수량')

# 수평막대 그래프
plt.subplot(1,2,2)
plt.barh(fruits, counts, height = 0.5) # (대상 y, 수량 x, hight)
plt.title('수평막대그래프')
plt.xlabel('수량')
plt.ylabel('과일')

plt.show()
```

## 3.2 막대그래프 활용

```python
# 2009년 계절별 강수량 비교

plt.bar(df.index, df[2009])
plt.grid(True)
plt.show()
```

- 막대 위에 수량값을 text로 출력
    - x좌표 : 0, 1, 2, 3
    - y좌표 : 수량

```python
# 막대 위에 수량값을 text로 출력
# x 좌표 기준으로 써지기에 문자열에 따라 중간으로 맞춰주기.

plt.figure(figsize = (7,7))
plt.bar(df.index, df[2009])


plt.text(1, 400, '안녕')


plt.grid(True)
plt.show()
```

```python
# 데이터레이블 
# x좌표 y좌표의 값은 x축과 y축의 tick 간격 비율에 따라 다르다.

plt.figure(figsize = (7,7))
plt.bar(df.index, df[2009])

for i, j in enumerate(df[2009]):
    plt.text(i-0.1, j+10, j)
    
# y축의 값의 범위를 지정    
plt.ylim(0, 850)

# x축의 값의 범위를 지정
# plt.xlim(-2, 10)
    
plt.grid(True)
plt.show()
```

```python
# 컬럼 별 평균
df.mean(axis=0)

# 인덱스 행 별 평균
round(df.mean(axis=1),2)
```

```python
# 계절별 평균 강수량을 비교
mean_rain = round(df.mean(axis=1),2) # mean_rain Series

# 대상 : mean_rain Series의 index,
# 수량 : mean_rain Seires의 values

plt.bar(mean_rain.index, mean_rain.values) 

plt.title('계절별 평균 강수량', fontsize = 20)
plt.xlabel('평균 강수량')
plt.ylabel('계절')

plt.grid(True)
plt.ylim(0,800)

for i,j in enumerate(mean_rain.values):
    plt.text(i-0.15, j+10, j)

plt.show()
```
