```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('data/서울시 연도별 황사 경보발령 현황.csv')
```

```python

df.rename(columns = {'최대농도(㎍/㎥/시)':'최대농도'}, inplace = True)
df
```

```python
fig = plt.figure(figsize= (20, 6))

# figure의 title
fig.suptitle('년도별 각종 변화량', fontsize = 25)

plt.subplot(1,2,1)
plt.plot(df['년도'], df['최대농도'], label = '최대농도')
plt.title('년도별 최대농도 변화량')
plt.xlabel('년도')
plt.ylabel('최대농도')


plt.subplot(1,2,2)
plt.plot(df['년도'], df['관측일수'], c = 'g')
plt.title('년도별 관측일수 변화량')
plt.xlabel('년도')
plt.ylabel('관측일수')

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize = (15, 10))

fig.suptitle('서울시 황사 통계량', fontsize = 25, c = 'g')


ax1 = fig.add_subplot(1,1,1)
ax2 = ax1.twinx()

ax1.plot(df['년도'], df['최대농도'], label='최대농도 변화량', c = 'b', alpha = 0.5)
ax2.plot(df['년도'], df['관측일수'], label='관측일수 변화량', c = 'pink', linestyle = '--')

ax1.set_xlabel('년도')
ax1.set_ylabel('최대농도')
ax2.set_ylabel('관측일수')

ax2.set_yticks([1,5,10,15, 20])

ax1.set_title('년도별 각종 변화량', fontsize = 20)

ax1.legend(bbox_to_anchor = (1.05,1), loc = 'upper left')
ax2.legend(bbox_to_anchor = (1.05,0.9), loc = 'upper left')



plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_excel('data/webbrowser_share.xls', converters = {'date':np.str})

df.set_index('date', inplace = True)
```

```python
df
```

```python
pd.options.display.max_columns = 30
df
```

```python
df.columns[0:7]
df[df.columns[7:]].sum(axis=1)
```

```python
new_df = df[df.columns[0:7]]
new_df["기타"] = df[df.columns[7:]].sum(axis=1)

new_df
```

```python
new_df
```

```python
new_df.index
```

```python
# 2019 년 상반기 1~6월 점유율 평균

plt.figure(figsize = (5,5))

m = new_df['2019.01':'2019.06'].mean()

plt.pie(m, labels = m.index, autopct = '%.2f%%', explode = [0.05,0.05,0,0,0,0,0,0],
       shadow = True, labeldistance = 1.6, pctdistance = 1.3)

plt.show()
```

```python
import pandas as pd

df = pd.read_excel('data/강수량.xlsx', index_col = '계절')
df
```

```python


plt.figure(figsize = (10,10))
plt.subplot(1,2,1)
plt.title('봄 강수량 변화량')
plt.bar(df.columns, df.loc['봄'], width = 0.7)
plt.grid(True)
plt.ylim(0,400)

plt.subplot(1,2,2)
plt.title('여름 강수량 변화량')
plt.barh(df.columns, df.loc['여름'], height = 0.3)

plt.grid(True)

plt.show()

```

```python
tips = pd.read_csv('data/tips.csv')
tips
```

```python


fig = plt.figure(figsize = (12, 12))
fig.suptitle('두 개의 분포도', fontsize = 17)

plt.subplot(2,1,1)
plt.hist(tips['total_bill'], bins = 100, alpha=0.5)
plt.xlabel('tip 금액')
plt.ylabel('tip 금액 별 빈도수')

plt.subplot(2,1,2)
plt.hist(tips['tip'], bins=100, color = 'r', alpha = 0.5)
plt.xlabel('tip 금액')
plt.ylabel('tip 금액 별 빈도수')

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize = (10, 8))

plt.boxplot(tips['tip'])
plt.show()
```

```python
plt.boxplot([tips['tip'], tips['total_bill']], labels = ['tip','total_bill'])
plt.show()
```
