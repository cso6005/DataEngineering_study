```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('data/    .csv')
```

```python

df.rename(columns = {'(//)':''}, inplace = True)
df
```

```python
fig = plt.figure(figsize= (20, 6))

# figure title
fig.suptitle('  ', fontsize = 25)

plt.subplot(1,2,1)
plt.plot(df[''], df[''], label = '')
plt.title('  ')
plt.xlabel('')
plt.ylabel('')


plt.subplot(1,2,2)
plt.plot(df[''], df[''], c = 'g')
plt.title('  ')
plt.xlabel('')
plt.ylabel('')

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize = (15, 10))

fig.suptitle('  ', fontsize = 25, c = 'g')


ax1 = fig.add_subplot(1,1,1)
ax2 = ax1.twinx()

ax1.plot(df[''], df[''], label=' ', c = 'b', alpha = 0.5)
ax2.plot(df[''], df[''], label=' ', c = 'pink', linestyle = '--')

ax1.set_xlabel('')
ax1.set_ylabel('')
ax2.set_ylabel('')

ax2.set_yticks([1,5,10,15, 20])

ax1.set_title('  ', fontsize = 20)

ax1.legend(bbox_to_anchor = (1.05,1), loc = 'upper left')
ax2.legend(bbox_to_anchor = (1.05,0.9), loc = 'upper left')



plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_excel('data/webbrowser_share.xls', converters = {'date':np.str})

df.set_index('date', inplace = True)
```

```python
df
```

```python
pd.options.display.max_columns = 30
df
```

```python
df.columns[0:7]
df[df.columns[7:]].sum(axis=1)
```

```python
new_df = df[df.columns[0:7]]
new_df[""] = df[df.columns[7:]].sum(axis=1)

new_df
```

```python
new_df
```

```python
new_df.index
```

```python
# 2019   1~6  

plt.figure(figsize = (5,5))

m = new_df['2019.01':'2019.06'].mean()

plt.pie(m, labels = m.index, autopct = '%.2f%%', explode = [0.05,0.05,0,0,0,0,0,0],
       shadow = True, labeldistance = 1.6, pctdistance = 1.3)

plt.show()
```

```python
import pandas as pd

df = pd.read_excel('data/.xlsx', index_col = '')
df
```

```python


plt.figure(figsize = (10,10))
plt.subplot(1,2,1)
plt.title('  ')
plt.bar(df.columns, df.loc[''], width = 0.7)
plt.grid(True)
plt.ylim(0,400)

plt.subplot(1,2,2)
plt.title('  ')
plt.barh(df.columns, df.loc[''], height = 0.3)

plt.grid(True)

plt.show()

```

```python
tips = pd.read_csv('data/tips.csv')
tips
```

```python


fig = plt.figure(figsize = (12, 12))
fig.suptitle('  ', fontsize = 17)

plt.subplot(2,1,1)
plt.hist(tips['total_bill'], bins = 100, alpha=0.5)
plt.xlabel('tip ')
plt.ylabel('tip   ')

plt.subplot(2,1,2)
plt.hist(tips['tip'], bins=100, color = 'r', alpha = 0.5)
plt.xlabel('tip ')
plt.ylabel('tip   ')

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize = (10, 8))

plt.boxplot(tips['tip'])
plt.show()
```

```python
plt.boxplot([tips['tip'], tips['total_bill']], labels = ['tip','total_bill'])
plt.show()
```
