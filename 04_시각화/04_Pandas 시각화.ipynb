## plot() 
- kind 매개변수에 지정한 값에 따라 다양한 그래프를 그릴 수 있다.
- 모든 그래프를 지정하는 것은 아니고 아래 그래프를 지원한다.
- kind : 그래프 종류 지정
    - 'line' : line plot (default)
    - 'bar' : vertical bar plot
    - 'barh' : horizontal bar plot
    - 'hist' : histogram
    - 'box' : boxplot
    - 'kde' : Kernel Density Estimation plot
    - 'pie' : pie plot
    - 'scatter' : scatter plot

### pibot_table로 정렬 후에 막대그래프로 그리기

- index : 수량을 셀 groupby 컬럼 (x축 대상 대분류)
- columns : 막대그래프를 나눠서 볼 groupby컬럼 (대상대분류 내 소분류막대)

## Boxplot (상자그래프)

- 극단치 : 정상범위를 넘어선 값들. 이상치(outlier) 중에 하나이다.

- 이상치 : 극단치, 오류값 -> 이상치라면 처리해줘야 한다. 오류값은 제거해줘야 함.

- 정상범위의 기준은?
- boxplot에서 극단치를 계산하는 방법
1. IQR(Inter Quantile Range) 계산
    - (3분위값 - 1분위값) : (중간 50% : 25~ 75% 의 크기)
2. 정상범위 이하의 극단치(극단적으로 작은 값) : (1분위값) - (IQR * 1.5)  보다 작은 값들
3. 정상범위 이상의 극단치(극단적으로 큰 값) : (3분위값) + (IQR * 1.5) 보다 큰 값
    - 계산한 정상범위 내에서 정상범위보다 실제 값 범위 최대, 최소이 작다면 그 최대, 최소에 선을 긋는다. 

- Whis 값 = (기본) 1.5 : 조절할 수 있는 값
    - 정상범위를 늘리고 싶은 경우 1.5 보다 큰 값을 곱하면 된다.
    - 정상범위를 줄이고 싶은 경우 1.5 보다 작은 값을 곱하면 된다.

```python
quantile= tips['total_bill'].quantile(q = [0.25, 0.5, 0.75])
IQR = quantile[0.75] - quantile[0.25]
whis = 1.5
o1 = quantile[0.25] - IQR * whis
o2 = quantile[0.75] + IQR * whis

print(f'정상범위: {o1} ~ {o2}')

tips['total_bill'][(tips['total_bill'] < o1) | (tips['total_bill'] > o2)]
```

```python
tips['total_bill'].plot(kind = 'box')
plt.show()
```

```python
# whis 변화에 따른 이상치 변화
# 하나의 figure에 여러 개의 axes를 사용하는 경우

fig = plt.figure(figsize = (15, 5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

# ax1에 그려라.
# 기본 1.5
tips['total_bill'].plot(kind = 'box', title='whis=1.5(기본)', ax = ax1) 

# ax2에 그려라.
# 1.5 -> 3 (늘림) : 정상범위가 넓어진다.
tips['total_bill'].plot.box(whis = 3, title='whis=3', ax = ax2)

# ax3에 그려라.
# 1.5 -> 1 (줄임) : 정상범위가 좁아진다.
tips['total_bill'].plot(kind = 'box', whis = 1, title= 'whis=1', ax = ax3)
```

```python
# whis 변화에 따른 이상치 변화 (plt)

plt.boxplot(tips['total_bill'], whis = 1)
plt.show()
```

```python
# 데이터프레임을 box plot하면 # DataFrame.plot(kind = 'box')
# 한 Axes 에 df의 컴럼별로 boxplot을 그린다. 

tips[['total_bill', 'tip', 'size']].plot(kind='box', figsize = (7,6))
plt.show()
```
