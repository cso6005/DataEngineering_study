> ### str accessor
> - `Series.str.String메소드()`
> - 문자열 타입 Series에 String 메소드들을 사용하여 일괄처리 할 수 있도록 한다.
> - str accessor 메소드 : https://pandas.pydata.org/pandas-docs/stable/reference/series.html#api-series-str

> ### datetime으로 변환
> - `pd.to_datetime(변환할 Series, format="날짜형식")`
> - x 또는 y 축이 날짜 일 경우 datetime 타입으로 처리하는 것이 좋다.
> - format: 년월일의 형식을 strftime 형식문자열로 지정한다. 
>      - [strftime 형식](https://docs.python.org/3/library/datetime.html#strftime-and-strptime-format-codes)  
> - datetime을 변환하면 날짜만의 기능을 다룰 수 있다.
>     - dt accessor를 이용해 datetime의 속성값들을(년, 월, 일 등) 추출할 수 있다.
>     - https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.dt.html


```python
# 지역 컬럼의 값을 이용하여 "구" 컬럼 생성 (지역에서 구만 추출)
gdf['지역'].apply(lambda x:x.split(' ')[1])
gdf['지역'].apply(lambda x:x[3:])
```

```python
# 지역 컬럼의 값을 이용하여 "구" 컬럼 생성 (지역에서 구만 추출)
gdf['구'] = gdf['지역'].str.slice(start = 3)
gdf
```

```python
# 기간 컬럼을 datetime 타입으로 변경
gdf['기간'] = pd.to_datetime(gdf['기간'], format = '%Y%m%d')
gdf.info()
```

```python
# 기간 컬럼을 이용해 "월", "일", "요일" 컬럼 생성
gdf['월'] = gdf['기간'].dt.month
gdf['일'] = gdf['기간'].dt.day
gdf['요일'] = gdf['기간'].dt.weekday
```

```python
# 요일 - 0:월, 1:화, 2:수, 3:목, 4:금, 5:토, 6:일 로 변경
def change_weekday(x):
    if x==0:
        return "월"
    elif x==1:
        return "화"
    elif x==2:
        return "수"
    elif x==3:
        return "목"
    elif x==4:
        return "금"
    elif x==5:
        return "토"
    elif x==6:
        return "일"
    
gdf['요일'].apply(change_weekday)
```

```python
gdf['요일'].apply(lambda x:"월화수목금토일"[x])
```

```python
# 요일 - 0:월, 1:화, 2:수, 3:목, 4:금, 5:토, 6:일 로 변경
week = pd.Series(['월', '화', '수', '목', '금', '토', '일'])
gdf['요일'] = week[gdf['요일']].values
gdf
```

```python
# 가장 휘발유가격이 저렴한 주유소 상위 5 - 같은 주유소가 여러개 나오므로 평균 또는 max값으로 (평균 기준)
싼주유소 = gdf.groupby('상호')['휘발유'].mean().sort_values().head()

# gdf.groupby('상호')['휘발유'].max().sort_values().head()
```

```python
# 휘발유 비싼 주유소의 상위 5개의 상호, 지역, 주소 조회(중복제거-drop_duplicates)
# 비싼 주유소 Series의 index를 이용해 조회

gdf.loc[gdf['상호'].isin(비싼주유소.index), ['상호', '지역', '주소']].drop_duplicates()
```

```python
gdf['상호'].isin(비싼주유소.index)
```

```python
# 휘발유 비싼 주유소의 상위 5개의 상호, 지역, 주소 조회(중복제거-drop_duplicates)

def check_max(x, y):
    return x['휘발유'].mean() >= y 

result = gdf.groupby('상호')['휘발유'].mean().sort_values(ascending = False).head()[4]
gdf3 = gdf.groupby('상호').filter(check_max, y = result)

gdf3.drop_duplicates(['상호'])[['상호', '지역', '주소']]
```

```python
### 휘발유 저렴한 주유소의 상위 5개의 상호, 지역, 주소 조회(중복제거)

def check_min(x, y):
    return x['휘발유'].mean() <= y 

result = gdf.groupby('상호')['휘발유'].mean().sort_values().head()[4]

gdf4 = gdf.groupby('상호').filter(check_min, y = result)

gdf4.drop_duplicates(['상호'])[['상호', '지역', '주소']]
```

```python
# 휘발유 저렴한 주유소의 상위 5개의 상호, 지역, 주소 조회(중복제거)


gdf.loc[gdf['상호'].isin(싼주유소.index), ['상호', '지역', '주소']].drop_duplicates()
```
