##### MLP(Multi Layer Perceptron) - 가장 기본적인 딥러닝모델의 구조

# Keras 개발 Process

1. 입력 텐서(X)와 출력 텐서(y)로 이뤄진 **훈련 데이터를 정의**

2. 입력과 출력을 연결하는 Layer(층)으로 이뤄진 **네트워크(모델) 구조를 을 정의** (딥러닝은 머신러닝과 달리 모델 구조는 우리가 정한다.) 
    - Sequential 방식: 순서대로 쌓아올린 네트워크로 이뤄진 모델을 생성하는 방식
    - Functional API 방식: 다양한 구조의 네트워크로 이뤄진 모델을 생성하는 방식
    - Subclass 방식: 네트워크를 정의하는 클래스를 구현.
    

3. **모델 Compile(컴파일)** - 모델을 학습할 수 있는 상태로 만들어준다.
    - 모델이 Train(학습)할때 사용할 손실함수(Loss Function), 최적화기법(Optimizer), 학습과정을 모니터링할 평가지표(Metrics)를 설정
    
    > Compile: 실행할 수 있는 상태(학습할 수 있는 상태)로 만들어 주는 것. (기계어로 바꾸는 의미 아님.)

4. **Training(학습/훈련)**
    - Train dataset을 이용해 모델을 Train 시킨다.

##### import

```python
import numpy as np

import tensorflow as tf # 별칭 : tf
from tensorflow import keras
```

## 네트워크(모델) 구현
- Network : 전체 모델 구조

```python
# Sequential(순차) 방식으로 모델 생성 
# layer : 단계
# input layer => HIDDEN 단계 - 해당 layer 단계에서 해야할 일을 하고 출력 내보내고 다음 layer 입력으로 들어가 일하고 ~~ = >마지막 최종 출력
# 이전 layer 의 출력이 현재 일할 layer의 입력.

model = keras.Sequential()

# 모델에 layer들을 추가 - model.add(레이어 객체)
# layer - 모델이 추론하는 각각의 단계를 의미
#    => Input(학습/추론 할 데이터를 넣어주는 단계), Hidde(추론과정을 담당), Output(추론결과를 출력) Layer 로 구성된다. 

# 1. Input Layer 추가
model.add(keras.layers.InputLayer((28,28))) 

# 2. Hidden Layer 들을 추가
# art : 모델 네트워크 만드는 것(내 데이터에 맞게 모델 구조를) - 성능에 영향을 주는
# 정해진 규칙이 있는게 아니라 내가 만드는 것
model.add(keras.layers.Flatten())  

model.add(keras.layers.Dense(256, activation='relu')) # Fully Connected Layer  # unit 256 (LR 256- 추론하는애가 256개)
model.add(keras.layers.Dense(128, activation='relu'))

# 3. Output Layer 추가
# 마지막에 추가된 layer는 output layer 이다.
model.add(keras.layers.Dense(10, activation='softmax'))
```

```python
# 모델의 구조를 확인(Text로 리턴)
model.summary() 

# 해당 layer 처리 후 결과 output 의 shape 를 말해줌. 
# 데이터의 갯수는 모르기에 none 이다. 이 정도만 알면 됨.
# 1차원 배열이 출력
# flatten 의 out 으로 (None, 784) 이 나와 dense 의 입력으로 들어간다.
# Param : 학습할 대상 파라미터 갯수- 업데이트 해야할 파라미터 개수. 
#         구조그림에서 선의 갯수 
```

```python
# pip install pydot pydotplus graphviz
# 모델의 구조를 확인(graphviz 이용해 시각화)
from tensorflow.keras.utils import plot_model
plot_model(model, to_file='model_shapes.png', show_shapes=True)
```

## 컴파일 단계 (학습에 필요한 기능을 추가)
- 정의된 네트워크 모델에 학습을 위한 추가 설정을 되어야 컴파일(학습이 가능한 상태)이 가능하다.
    - Optimizer (필수) : 최적화 알고리즘
    - 손실함수 (필수) 
    - 평가지표

```python
model.compile(optimizer='adam', # 최적화 함수 : Adam 알고리즘 
              loss='categorical_crossentropy', # cross entropy
              metrics=['accuracy']) # 추가 평가지표 - 정확도
```

## 데이터 준비
- X (Input Data Image)
    - 0 ~ 255 = > 0 ~ 1 = > 기존값/255 # minmaxscale 할 필요 없음.(그건 feature 마다 다를 때 인데 여기서는 255 동일하기에)
    - 0 ~ 1 사이의 값으로 정규화 시킨다.
- y (Output Data) 0 ~ 9
    - one hot encoding 처리
        - Label이 다중분류(Multi class classification)일 경우 One Hot Encoding 한다.
    - Keras의 onehot encoding 처리 함수
        - tensorflow.keras.utils.to_categorical()
        
- 일반적인 8 bit 정수표현  : signed int 최상위비트 부호 비트. -128 ~ 127
- 색농도 0 ~ 255표현 : 음수필요 없음. unsigned int

```python
# uint8 : unsigned int - 8bit
print(np.min(train_image), np.max(train_image), train_image.dtype, sep=', ')
```

##### input image(X)를 정규화. 
- 이미지를 0 ~ 1 로 정규화 하고 타입을 float32로 변환

```python
X_train = train_image.astype(np.float32) # 타입 변환: unit8 -> float32
X_train /= 255.0 # X_train= X_train/255.0 # X_train 이 가지는 제일 큰 값(max)인 255로 나누면  0 ~ 1 로 정규화.

X_test = test_image.astype(np.float32)
X_test /= 255.0
```

```python
print(np.min(X_train), np.max(X_test), X_train.dtype, sep=', ')
```

##### label(y)를 one hot encoding
- 모델이 각 클래스 별 확률이 출력되도록 정의 되었기 때문에 학습시(최적화시) 오차를 구하기 위해서는 정답이 label 이 아니라 확률이 되도록 y 를 one hot encoding 
- 모델이 출력을 하는데 마치 sklearn 에서 보면 predict proba 처럼
- 오차까지 보기 위해 label의 확률을 보려고 one hot encoding 한다. ex. 0 일 확률은 ~.


- 모든 머신러닝, 딥너링에서 제공한다. 
    - keras.utils.to_categorical()

```python
print(train_label.shape, test_label.shape)
train_label[:5]
```

```python
np.unique(train_label)
```

```python

y_train = keras.utils.to_categorical(train_label)
y_test = keras.utils.to_categorical(test_label)
```

```python
y_train[:5]
```

```python
print(train_label.shape, test_label.shape)
print(y_train.shape, y_test.shape)
```

## 학습 (fit)
- model.fit()
    - 모델 학습 메소드. 
    - 학습과정의 Log를 History 객체로 반환한다.
- History : train 시 에폭별 평가지표값들을 모아서 제공.    

- epoch: 10
- batch_size = 100


- 문제집(train set) 에 문제가 60,000개 
- 1 epoch : 6만개 문제를 한번 다 풀어보는 것.
- 100 문제(batch size) : 풀고 정답확인하고 다음 100문제 풀고 정답확인하고 ... 백개의 단위로 문제를 푼다.
- batch_size 단위로 학습 = > 1step 이라고 이야기한다.

```python
# hist : 모델이 학습끝나면 epoch 별 train 오차loss - accuracy - validation 오차(loss) - validation 정확도 를 반환
# 학습한 history를 변수에 담아준다.
# epoch 별 history 를 시각화해서 본다.

hist = model.fit(X_train, y_train, # input: X_train output : y_train
                  epochs=10,  # 전체 train dataset 을 몇 번 반복해서 학습시킬지 정의. - 1 epoch (단위개념) : 한 번 학습 // 데이터를 여러번 봐야할 필요가 있다. 몇 백 몇 천 줄 수 도 있음. 
                  batch_size=100, # Train dataset 을 학습시키는 단위로 한번에 지정한 개수(100)의 데이터 씩 끊어서 학습시킨다. batch-size를 한번 학습한 것을 1 step(단위) 
                  validation_split=0.3 # train/validation 분리 - train set 의 30%(0.3)를 validation set으로 사용 
                 )

# train set 42000

# Epoch 1/10 
# 현재 학습중인 epoch/ 전체 epoch수

# 420/420 [==============================] - 4s 8ms/step - loss: 0.3048 - accuracy: 0.9122 - val_loss: 0.1568 - val_accuracy: 0.9543
# 현재 학습중인 step/전체 step 수          # 1epoch 걸린시간:4s 1step 학습에 걸린시간: 8ms  # - train 오차loss - accuracy - validation 오차(loss) - validation 정확도 # 원래 loss 나오고 아까 추가 평가지표 정확도 추가해서

# train 70 % 가지고 학습하고 train 과 val set 으로 검증을 한다. 

# train 은 여러번할수록 좋아진다. # test은 좋아지다가 안 좋아짐 => 모델이 복잡해짐. overfitting
```

##### History 시각화

```python
print(type(hist))
```

```python
# hist.history - {key 성능이름 :epoch 별 성능변화}
print(type(hist.history))
hist.history
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(7,10))

plt.subplot(2,1,1)
plt.plot(range(1,11), hist.history['loss'], label='Train Loss')
plt.plot(range(1,11), hist.history['val_loss'], label="Validation Loss")
plt.title("Loss per Epoch", fontsize=20)
plt.legend()

plt.subplot(2,1,2)
plt.plot(range(1,11), hist.history['accuracy'], label='Train Accuracy')
plt.plot(range(1,11), hist.history['val_accuracy'], label='Validation Accuracy')
plt.title("Accuracy per Epoch", fontsize=20)

plt.legend()
plt.tight_layout()
plt.show()

# loss 오차 - 작을수록 좋은 것. 
# accuracy 정확도 - 클수록 좋은 것.

# 3 epoch 5 epoch 에서 성능 좋음.
```

## 테스트셋 평가 - > 최종평가
- model.evaluate(X,y)

```python
test_loss, test_acc = model.evaluate(X_test, y_test)
```

```python
print(test_loss, test_acc)
```

## 새로운 데이터 추론
- 새로운 데이터를 추론하기 전에 학습데이터에 했던 전처리과정을 동일하게 적용 한 뒤 추론한다.


- input image(X) : 28x28,  float32, 0 ~ 1 정규화, 검은색바탕에 흰 글씨

### 추론 메소드
- predict() - 회귀이든 분류이든
    - **분류:** 각 클래스 별 확률 반환
    - **회귀:** 최종 예측 결과


- 분류문제일때 predict() 결과에서 class label 출력하기 = > 모델 출력값을 후처리해서 최종 예측 label 을 확인한다.   
    - 이진 분류(binary classification)
        - `numpy.where(model.predict(x) > 0.5, 1, 0).astype("int32")`
    - 다중클래스 분류(multi-class classification)
        - `numpy.argmax(model.predict(x), axis=1)`

```python
# X_test 의 처음 5개 데이터를 이용해 새로운 데이터라 하고 추론
# X_test 는 test_image 를 전처리한 데이터셋(unit8 ->float32 변환, 0 ~ 1 사이로 정규화) - 해줄 필요 없음.

new_image = X_test[:5] # 28x28 이미지 5개
new_image.shape
```

```python
# predict(예측할 데이터의 Feature) -> (데이터개수, Feature shape)
# (5,28,28) -> (5,10) : 5개 이미지에 대해 0~9 (onehotencoding)
result = model.predict(new_image)
result.shape
```

```python
result[0] # 0 번 데이터에 대한 예측결과(10,) : 0 ~ 9 일 각각의 확률! 반환. 
```

```python
# 가장 높은 확률
print(np.max(result[0]))

# 가장 높은 확률이 있는 index
np.argmax(result, axis=-1) # --> (5, 10)에서 수평 방향으로 10개 중에 제일 큰 것. => axis = -1 :마지막 축 기준으로 max 값을 가진 index 를 조회 (5,10) 마지막 축은 1번축
```

```python
# label 을 확인.
result_label = np.argmax(result, axis=-1)
```
