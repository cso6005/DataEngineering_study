### fasting blood sugar에 따른 흡연 유무

- 정상: 70~100mg/dl, 60대이상: 110mg/dl이하

```python
smoking.insert(13,'20대 이상 당뇨',(smoking['fasting blood sugar']>100) & (smoking['age']<60))
```

```python
smoking[['fasting blood sugar','20대 이상 당뇨','60대 이상 당뇨']]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler, Binarizer
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import (confusion_matrix, plot_confusion_matrix, ConfusionMatrixDisplay, accuracy_score,
                             recall_score, precision_score, f1_score, precision_recall_curve,roc_auc_score, plot_roc_curve,
                             average_precision_score, classification_report
                             )
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression

import pandas as pd
import numpy as np
```

```python
X = smoking.drop('smoking',axis=1)
y = smoking['smoking']

X.shape,y.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0,test_size=0.2)
X_train,X_val,y_train,y_val=train_test_split(X_train,y_train,random_state=0,test_size=0.2)
```

### LabelEncoder

```python
lb = LabelEncoder()
X_train['gender'] = lb.fit_transform(X_train['gender'])
X_test['gender'] = lb.transform(X_test['gender'])
```

```python
lb = LabelEncoder()
X_val['gender'] = lb.fit_transform(X_val['gender'])
```

```python
lb = LabelEncoder()
X_train['tartar'] = lb.fit_transform(X_train['tartar'])
X_test['tartar'] = lb.transform(X_test['tartar'])
```

```python
lb = LabelEncoder()
X_val['tartar'] = lb.fit_transform(X_val['tartar'])
```

```python
lb = LabelEncoder()
X_train['oral'] = lb.fit_transform(X_train['oral'])
X_test['oral'] = lb.transform(X_test['oral'])
```

```python
lb = LabelEncoder()
X_val['oral'] = lb.fit_transform(X_val['oral'])
```

### StandardScaler

```python
st_s = StandardScaler()
st_s.fit(X_train)
X_train= st_s.transform(X_train)  
X_test = st_s.transform(X_test)
```

```python
X_val = st_s.transform(X_val)
```

### DecisionTreeClassifier

```python
params = {
    'max_depth':range(1,30),
    'max_leaf_nodes':range(10,101,5),
    'min_samples_leaf':range(1,501,50)
}
gs = GridSearchCV(DecisionTreeClassifier(random_state=0),
                  params,
                  scoring='accuracy',
                  cv=5,
                  n_jobs=-1)

```

```python
gs.fit(X_train,y_train)
```

### 결과

```python
gs.cv_results_
```

```python
gs_result = pd.DataFrame(gs.cv_results_)
gs_result.sort_values('rank_test_score').head()
```

```python
print("가장 좋은 평가점수: ",gs.best_score_)
```

```python
print("가장 좋은 파라미터 조합:",gs.best_params_)
```

### best model

```python
gs_best_model = gs.best_estimator_
pred_val = gs_best_model.predict(X_val)
pred_test = gs_best_model.predict(X_test)
```

```python
# 분류 평가지표들을 묶어서 출력하는 함수.
def print_classification_metrics(y, pred, title=None):
    print('='*50)
    if title:
        print(title)
        print('-'*50)
    print('정확도(Accuracy):', accuracy_score(y, pred))
    print('재현율/민감도(Recall):', recall_score(y, pred))
    print('정밀도(Precision):', precision_score(y, pred))
    print('F1 점수(F1 Score):', f1_score(y, pred))
    print('='*50)
```

```python
print_classification_metrics(y_val,pred_val,title='val 결과')
print('\n')
print_classification_metrics(y_test,pred_test,title='test 결과')
```
